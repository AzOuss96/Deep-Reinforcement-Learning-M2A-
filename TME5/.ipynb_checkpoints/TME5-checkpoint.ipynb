```python
import matplotlib
matplotlib.use("TkAgg")
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
import gym
from gym import wrappers, logger
import numpy as np
import copy
import torch
from torch import nn
from torch import FloatTensor
from torch import *
import random
from torch.utils.tensorboard import SummaryWriter
from torch.distributions import Categorical
import torch.optim as optim
import os,sys,inspect

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)

from modules.utils import ReplayMemory, NeuralNet
from modules.A2C import A2CAgent
```

```python
################################
###   Online Learning   ########
################################


if __name__ == '__main__':

    ### hyper-parameters ###
    GAMMA = 0.99

    ### Environment setting ###
    env_name = 'CartPole-v1'
    #env_name = 'LunarLander-v2'
    env = gym.make(env_name)
    outdir = env_name + '/A2C-results'
    
    action_size = env.action_space.n 
    state_size = env.observation_space.shape[0]
    
    envm = wrappers.Monitor(env, directory=outdir, force=True, video_callable=False)
    env.seed(0)
    env.reset()
    done = False
    verbose = False
    
    batch_size = 10
    replay_mem = ReplayMemory(batch_size=batch_size) 
    
    layers_actor = [10, 10, 10]
    layers_critic = [10, 10, 10]

    ### Initialization ### Actor + Critic
    a2c = A2CAgent(actor = NeuralNet(state_size, action_size, layers_actor).double(),
                      critic = NeuralNet(state_size, 1, layers_critic, soft=False).double())


    ### Training Settings ###
    episode_count = 7000
    optimizerActor = optim.Adam(a2c.actor.parameters())
    optimizerCritic = optim.Adam(a2c.critic.parameters())
    huberLoss = nn.SmoothL1Loss()

    rsum_hist = []

    ### Training loop ###

    for episode in range(episode_count):

        it = 0
        obs = envm.reset()
        rsum = 0.0
        done = False

        while(True):

            if verbose == True:
                env.render()
            
            state = obs
            action = a2c.act(state)
            obs, reward, done, _ = envm.step(action.item())
            rsum += reward
            it += 1

            #### batch update ####  
            states, actions, rewards, obss, dones = state, action, reward, obs, done
            next_actions = a2c.act(obss)
            _, next_values, _ = a2c.evaluate(obss, next_actions)
            logprobs, values, dist_entropy  = a2c.evaluate(states, actions)

            ## evaluating advantages ##
            returns = rewards + GAMMA * next_values * (1.0 - torch.tensor(dones).double())
            advantages = returns - values

            ## critic loss ##
            optimizerCritic.zero_grad()
            critic_loss = huberLoss(returns, values)
            critic_loss.backward()
            optimizerCritic.step()

            ## actor loss ##
            optimizerActor.zero_grad()
            actor_loss = - (advantages.detach() * logprobs).mean()
            actor_loss.backward()
            optimizerActor.step()
            
            if done == True:
                print("Episode : " + str(episode) + " rsum=" + str(rsum), "iter = "+ str(it))
                rsum_hist.append(rsum)
                break
```

```python
############################
### Roll out Monte Carlo ###
############################

if __name__ == '__main__':

    ### hyper-parameters ###
    GAMMA = 0.99

    ### Environment setting ###
    env_name = 'CartPole-v1'
    #env_name = 'LunarLander-v2'
    env = gym.make(env_name)
    outdir = env_name + '/A2C-results'
    
    action_size = env.action_space.n 
    state_size = env.observation_space.shape[0]
    
    envm = wrappers.Monitor(env, directory=outdir, force=True, video_callable=False)
    env.seed(0)
    env.reset()
    done = False
    verbose = False
    
    batch_size = 10
    replay_mem = ReplayMemory(batch_size=batch_size) 
    
    layers_actor = [10, 10, 10]
    layers_critic = [10, 10, 10]

    ### Initialization ### Actor + Critic
    a2c = A2CAgent(actor = NeuralNet(state_size, action_size, layers_actor).double(),
                      critic = NeuralNet(state_size, 1, layers_critic, soft=False).double())


    ### Training Settings ###
    episode_count = 7000
    optimizerActor = optim.Adam(a2c.actor.parameters())
    optimizerCritic = optim.Adam(a2c.critic.parameters())
    huberLoss = nn.SmoothL1Loss()

    rsum_hist = []

    ### Training loop ###

    for episode in range(episode_count):

        it = 0
        obs = envm.reset()
        rsum = 0.0
        done = False

        while(True):

            if verbose == True:
                env.render()
            
            state = obs
            action = a2c.act(state)
            obs, reward, done, _ = envm.step(action.item())
            rsum += reward
            it += 1
            
            replay_mem.store(state, action, reward, obs, done)
            
            if done == True:
                print("Episode : " + str(episode) + " rsum=" + str(rsum), "iter = "+ str(it))
                rsum_hist.append(rsum)
                break
                
        #### batch update ####
        states, actions, rewards, obss, dones = replay_mem.pop(batch_size=len(replay_mem))    
        next_actions = a2c.act(obss)
        _, next_values, _ = a2c.evaluate(obss, next_actions)
        logprobs, values, dist_entropy  = a2c.evaluate(states, actions)

        ## evaluating advantages ##
        returns = rewards + GAMMA * next_values * (1.0 - dones.double())
        advantages = returns - values

        ## critic loss ##
        optimizerCritic.zero_grad()
        critic_loss = huberLoss(returns, values)
        critic_loss.backward()
        optimizerCritic.step()

        ## actor loss ##
        optimizerActor.zero_grad()
        actor_loss = - (advantages.detach() * logprobs).mean()
        actor_loss.backward()
        optimizerActor.step()
            

```

```python
env.close()
```

```python

window = 10
data = pd.concat([pd.Series(rsum_hist, name='mean').rolling(window).mean(),
          pd.Series(rsum_hist, name='std').rolling(window).std()],
          axis=1)

sns.set()
plt.figure(figsize=(10,5))
ax = data['mean'].plot()

ax.fill_between(data.index, data['mean'] - data['std'], data['mean'] + data['std'],
                alpha=.25)
plt.tight_layout()
plt.ylabel("Cumulated sum of rewards (Y)")
plt.title('%s :  $\gamma = %.2f$ | Roll out MC'%(env_name, GAMMA))
plt.legend(['$%d$-Episod rolling mean of Y'%window, '$%d$-Episod rolling std of Y'%window])
plt.xlabel('Episod')
sns.despine()
```
